```python
ruta = "../data/Exportaciones * 2024.txt"

indices = [0,14,20,69,73]
cols_nombres = ["FECHA","REGION_ID", "PAIS_DESTINO", "CODIGO_HS", "VALOR_FOB"]

mapa_regiones = {"1":"Tarapacá","2":"Antofagasta","3":"Atacama","4":"Coquimbo","5":"Coquimbo","6":"OHiggins","7":"Maule","8":"Biobio","9":"Araucania","10":"Los lagos",
                 "11":"Aysen","12":"Magallanes","13":"Metropolitana","14":"Los rios","15":"Arica y parinacota","16":"Ñuble","20":"Indeterminado"}

```

```python
archivos = glob.glob(ruta)
archivos.sort()

for a in archivos[:3]:
    print(os.path.basename(a))
    
lista_dfs = []

if len(archivos) == 0:
    print("error no hay archivos")
    
else:
    for archivo in archivos:
        try:
            df_temp = pd.read_csv(
                archivo,
                sep=";",
                encoding="latin1",
                decimal=",",
                header=None,
                usecols=indices,
                dtype={14: str, 69: str}
            )
            lista_dfs.append(df_temp)
        except Exception as e:
            print("error",archivo,e)
    
    if lista_dfs:
        df_final = pd.concat(lista_dfs,ignore_index=True)
        df_final.columns = cols_nombres
        
        df_final["NOMBRE_REGION"] = df_final["REGION_ID"].map(mapa_regiones).fillna("otro")
        df_final["CODIGO_HS"] = df_final["CODIGO_HS"].str.replace(".","",regex=False).str.strip()
        
        os.makedirs("../data/processed",exist_ok=True)
        df_final.to_parquet("../data/processed/chilex_2024.parquet",index = False)
        
```
